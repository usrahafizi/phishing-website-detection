# 2. Phishing Dataset Cleaning
This dataset have 89 featuers and 11430 rows. Each columns indecates a specific part of the link.

```python
# rename some columns 
phishing_dataset.rename(columns={
    'length_url':'length_url_in_nb',
    'status':'phishing'
}, inplace = True)
```

```python
# we extract the features that include URL in header like number of & in URL, 
# except the Phishing column which is the target.
for i in phishing_dataset.columns:
    if 'nb' not in i and i != 'phishing':
        phishing_dataset.drop(i, axis = 1, inplace = True)
phishing_dataset
        
```

```python
# see what columns we have
phishing_dataset.columns
```

```python
# replace all the -1 with 0 becauce it does not make sence
for k in phishing_dataset.columns:
    phishing_dataset[k].replace([-1], 0)
```

```python
# rename all (nb) to number to be more readable in the graph
# phishing_dataset.drop('phishing', axis = 1, inplace = True)
dct_1 = {}
dct_2 = {}
for n in phishing_dataset.columns:
    dct_1[str(n.split('_')[0]+ ' of ' + n.split('_')[1]).replace('nb','Number')] = sum(phishing_dataset[n])
    if n != 'length_url_in_nb':
        dct_2[str(n.split('_')[0]+ ' of ' + n.split('_')[1]).replace('nb','Number')] = sum(phishing_dataset[n])
```

```python
phishing_dataset
```

```python
# add the top 7 column, to new_dct to letter represent it in graph
plt.figure(figsize = (8,5))
temp = sorted(dct_1.items(), key = lambda x: x[1], reverse = True)
new_dct = {}
b = 0 
for c in temp:
    title,value = c
    if b<=4:
        new_dct[title] = value
        b+=1
new_dct = dict(sorted(new_dct.items(), key= lambda item: item[1], reverse = True))
new_dct = dict(itertools.islice(new_dct.items(), 7))

ax = sns.barplot(x = list(new_dct.values()),
                 y = list(new_dct.keys()))
ax.set_title('The features of most repetitive phishing webistes')
new_dct = dict(sorted(new_dct.items(), key = lambda item: item[1], reverse = True))
plt.show()
```
